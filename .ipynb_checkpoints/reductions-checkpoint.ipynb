```python
import glob
from wavescapes import single_wavescape, all_wavescapes, legend_decomposition
from matplotlib import pyplot as plt
```

```python
#producing all the plots
scores = glob.glob("./reductions/*.mid")

for score in scores:
    print(str(score.split('\\')[-1].split('.')[0]))
    plt.suptitle(str(score.split('\\')[-1].split('.')[0]) + '\n', fontsize=14)
    all_wavescapes(score, individual_width=500, aw_size=1.0, aw_per_tick=4, tick_start=1)
        
    plt.savefig('./reductions_figures/' + str(score.split('\\')[-1].split('.')[0]) + '.png')
```
