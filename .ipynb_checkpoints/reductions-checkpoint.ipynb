```python
import glob
from wavescapes import *
from matplotlib import pyplot as plt
```

```python
#producing all the plots
scores = glob.glob("./reductions/*.mid")

for score in scores:
    print(str(score.split('\\')[-1]))
    plt.suptitle(str(score.split('\\')[-1]) + '\n', fontsize=14)
    all_wavescapes(score, individual_width=500, aw_size=1.0, aw_per_tick=4, tick_start=1)
        
    plt.savefig('./reductions_figures/' + str(score.split('\\')[-1]) + '.png')
```

To better interpret the plots:
1. The distribution of magnitudes of the coefficients indicate which scales/chords are detected, as in 
![magnitude_1](figures/scales.png "Fig.1: Interpreting the magnitude of the coefficients")


2. The color of wavescapes indicates the 'tonic' (or dominant), as in 
![color_1](figures/colors.png "Fig.2: Interpreting the colors of the plot")
