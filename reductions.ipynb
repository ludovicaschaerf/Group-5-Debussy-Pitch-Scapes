```python
#producing all the plots
scores = glob.glob("./reductions/*.mid")

for score in scores:
    print(str(score.split('\\')[-1]))
    plt.suptitle(str(score.split('\\')[-1]) + '\n', fontsize=14)
    all_wavescapes(score, individual_width=500, aw_size=1.0, aw_per_tick=4, tick_start=1)
        
    plt.savefig('./reductions_figures/' + str(score.split('\\')[-1]) + '.png')
```
